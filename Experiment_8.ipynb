```python
import cv2
import numpy as np
import requests
from io import BytesIO
from google.colab.patches import cv2_imshow  # Import cv2_imshow for displaying images in Colab

# 1. URL of the image
image_url = 'https://tse4.mm.bing.net/th?id=OIP.-OPlR_eRVYNy5IbdXCW5rwHaDt&pid=Api&P=0&h=180'

# 2. Fetch the image from the URL
response = requests.get(image_url)

# Check if the request was successful
if response.status_code == 200:
    # Convert the image to a NumPy array
    image_array = np.asarray(bytearray(response.content), dtype=np.uint8)

    # Decode the image
    image = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

    if image is None:
        print("Error: Image could not be loaded from the URL.")
    else:
        # 3. Convert the image to grayscale
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Display the grayscale image in Colab
        cv2_imshow(gray_image)

        # 4. Create a white image (all pixel values set to 255)
        white_image = np.ones_like(image) * 255

        # Create a black image (all pixel values set to 0)
        black_image = np.zeros_like(image)

        # Display white and black images
        cv2_imshow(white_image)
        cv2_imshow(black_image)

        # 5. Draw shapes on an image
        # Draw a rectangle: (image, top-left corner, bottom-right corner, color, thickness)
        cv2.rectangle(image, (50, 50), (300, 300), (0, 255, 0), 5)

        # Draw a circle: (image, center, radius, color, thickness)
        cv2.circle(image, (400, 200), 50, (0, 0, 255), 5)

        # Draw a line: (image, start_point, end_point, color, thickness)
        cv2.line(image, (0, 0), (500, 500), (255, 0, 0), 5)

        # Display the image with shapes
        cv2_imshow(image)
```
